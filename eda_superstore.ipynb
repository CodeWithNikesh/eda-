```python
# droping unnecessary columns
data =   data.drop(['Row ID', 'Order ID','Customer ID','Postal Code'],axis = 1)
```

```python
# Technology products have the highest profit margin compared to other product categories

category_profit = data.groupby('Category')['Profit'].sum()
category_profit.plot(kind='bar')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Total profit')
plt.show()
# from the three category technology product have the higest profit margine
```

```python
region_sales = data.groupby('Region')['Sales'].sum()
region_sales.plot(kind='bar')
plt.title('Total region sales')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()

#  western region has the higest sales compare to other region
```

```python
#  sales are higher during the certain month of the year
data['Order Month'] = pd.DatetimeIndex(data['Order Date']).month
month_sales = data.groupby('Order Month')['Sales'].sum()
month_sales.plot(kind='line')
plt.title('Total sales by month')
plt.xlabel('Month')
plt.ylabel('Total sales')
plt.show()
#  sales are higher during the certain month of the year
```

```python
# order with the  same day shipping have the lowst rate of returned product
total_order_by_shipping_mode = data.groupby('Ship Mode').size()

returned_orders = data[data['Profit']<0].groupby('Ship Mode').size()

returned_orders_percentyage = (returned_orders/total_order_by_shipping_mode)*100
returned_orders_percentyage.plot(kind='bar')
plt.title('Return percentage by shipping mode')
plt.xlabel('Shipping Mode')
plt.ylabel('Return percentage')
plt.show()

#  standard class has the highest return percentage
```

```python
data['Order Day'] =pd.DatetimeIndex(data['Order Date']).day_name()
day_profit = data.groupby('Order Day')['Profit'].sum()
day_profit.plot(kind='bar')
plt.title('Total profit by the day of the week')
plt.xlabel('Days')
plt.ylabel('Profit')
plt.show()
```
